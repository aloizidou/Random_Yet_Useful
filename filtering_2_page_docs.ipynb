```python
import os
import shutil
import pdfplumber
```

```python
base_dir = "/Users/andrealoizidou/Desktop/capstone_project/authentic_resumes"
output_dir = os.path.join(base_dir, "filtered_2_page_resumes")
os.makedirs(output_dir, exist_ok=True)

# Iterate through all subdirectories
for subdir in os.listdir(base_dir):
    subdir_path = os.path.join(base_dir, subdir)
    
    # Ensure it's a directory
    if os.path.isdir(subdir_path):
        # Iterate through all PDF files
        for filename in os.listdir(subdir_path):
            if filename.endswith(".pdf"):  
                file_path = os.path.join(subdir_path, filename)
                
                # Count pages in the PDF
                try:
                    with pdfplumber.open(file_path) as pdf:
                        if len(pdf.pages) == 2:  # exactly 2 pages
                            shutil.copy(file_path, output_dir)  # copy to new folder
                            print(f"Copied: {filename}")
                except Exception as e:
                    print(f"Error processing {filename}: {e}")

print("Filtering complete. Check the 'filtered_2_page_resumes' folder for results.")
```
